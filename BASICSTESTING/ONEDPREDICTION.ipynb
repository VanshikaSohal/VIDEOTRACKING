```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#TRUE POSITION OF OBJECT(moving linearly)
true_position=np.linspace(0,50,50)
```

```python
#STIMULATED NOISY MEASUREMENTS
measurement=true_position+np.random.normal(0,2,true_position.shape)
```

```python
#Kalman Filter Variables
x=0.0 # initial estimate
P=1.0  # initial uncertainty
Q=1e-5 # process noise
R=4.0   # measurement noise
K_values=[]
predictions=[]
for z in measurement:
    #PREDICTION STEP
    x=x
    P=P+Q
    #UPDATE STEPS
    K=P/(P+R)
    x=x+K*(z-x)
    P=(1-K)*P
```

```python
#STORE RESULTS
predictions.append(x)
K_values.append(K)
```

```python
#PLOT RESULTS
plt.plot(true_position,label='True Position')
plt.plot(measurement,'r.',label='Measurements')
plt.plot(predictions,'g-',label='Kalman Filter Prediction')
plt.legend()
plt.xlabel('Time step')
plt.ylabel('Position')
plt.title('1D Kalman Filter Example')
plt.show()
```
